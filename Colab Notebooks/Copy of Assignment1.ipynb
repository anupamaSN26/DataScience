```python
# import numpy, pandas, matplotlib,seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#load the dataset "students" (posted in the classroom)
df = pd.read_csv("/content/students (4).csv")
#display first 10 rows
df.head(10)

```

```python
#Plot line plot of the field FinalMark label the graph, line style is dotted and color is green

plt.plot(df['FinalMark'], linestyle='dotted', color='green')
plt.title('FinalMark Line Plot')
plt.show()
```

```python
#draw scatter plot showing the relation Series1 and FinalMark, Series2 and Final Marks . Give labels to axes and legend

df['Sum_Series1_Series2'] = df['Series1'] + df['Series2']
plt.scatter(df['Sum_Series1_Series2'], df['FinalMark'])
plt.xlabel('Sum of Series1 and Series2')
plt.ylabel('FinalMark')
plt.show()
```

```python
#Box Plot of FinalMarks and find the median

plt.boxplot(df['FinalMark'])
plt.ylabel('FinalMark')
median = df['FinalMark'].median()
print(f'Median of FinalMark: {median}')
plt.show()
```

```python
#import seaborn and draw boxplot
sns.boxplot(y=df['Series1'])
plt.ylabel('Series1')
percentile_25 = df['Series1'].quantile(0.25)
lowest_25_highest = df[df['Series1'] <= percentile_25]['Series1'].max()
print(f'Highest Mark obtained by the lowest 25% Students in Series1: {lowest_25_highest}')
plt.show()
sns.boxplot(y=df['Series2'])
plt.ylabel('Series2')
percentile_75 = df['Series2'].quantile(0.75)
highest_75_highest = df[df['Series2'] <= percentile_75]['Series2'].max()
print(f'Highest Mark obtained by the 75% Students in Series2: {highest_75_highest}')
plt.show()

```

Highest Mark obtained by the lowest 25% Students in Series1

Highest Mark obtained by the 75% Students in Series2

Ans:18.2
