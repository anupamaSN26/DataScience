Link of dataset :https://raw.githubusercontent.com/smitha-ks/Notebooks/Datasets/diabetes_dataset_DT.csv

Use diabetes dataset to implement decision tree algorithm

```python
# Import the necessary libraries
import  pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report


```

```python
# Load dataset

#df=pd.read_csv("/content/diabetes_dataset_DT.csv")
diabetes=datasets.load_diabetes()
x=diabetes.data
y=diabetes.target
```

```python
# Split the dataset into a training set and a testing set(trainig 66% records)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.34,random_state=1)
```

```python

# Create a Decision Tree Classifier with gini index as criterion

DT_clf=DecisionTreeClassifier(criterion="gini",random_state=1)
```

```python
# Train the classifier on the training data
DT_clf.fit(x_train,y_train)
```

```python

# Make predictions on the test data
y_pred=DT_clf.predict(x_test)

# Calculate accuracy and confusion matrix
accuracy=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
cf=classification_report(y_test,y_pred)
print("Accuracy score: ",accuracy)
print("Confusion matrix : \n",cm)
print("Classification report: \n",cf)

```

```python
#Plot decision tree

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(15,10))
plot_tree(DT_clf)
plt.show()
```
